```python
def start_request(self):
    urls=[]

    for pids in train_pid_data:
        urls.append(f'https://eutils.ncbi.nlm.nih.gov/entrez/eutils/\
        efetch.fcgi?db=pubmed&id={pids}&rettype=xml&retmode=xml')

    for url in urls:
        yield scrapy.Request(url=url, callback=parse)
```
